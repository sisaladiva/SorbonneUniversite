```python
import subprocess

class OpensslError(Exception):
    pass

class KeyPair:
    def __init__(self, private_key_path, public_key_path):
        self.private_key_path = private_key_path
        self.public_key_path = public_key_path

    def encrypt(self, plaintext, cipher='aes-128-cbc'):
        # Use the public key to encrypt the plaintext
        args = ['openssl', 'rsautl', '-encrypt', '-' + cipher, '-pubin', '-inkey', self.public_key_path]
        result = subprocess.run(args, input=plaintext.encode('utf-8'), stdout=subprocess.PIPE, stderr=subprocess.PIPE)

        # If there is an error message on stderr, raise OpensslError
        error_message = result.stderr.decode()
        if error_message != '':
            raise OpensslError(error_message)

        # Return the encrypted content
        return result.stdout

    def decrypt(self, ciphertext, cipher='aes-128-cbc'):
        # Use the private key to decrypt the ciphertext
        args = ['openssl', 'rsautl', '-decrypt', '-' + cipher, '-inkey', self.private_key_path]
        result = subprocess.run(args, input=ciphertext, stdout=subprocess.PIPE, stderr=subprocess.PIPE)

        # If there is an error message on stderr, raise OpensslError
        error_message = result.stderr.decode()
        if error_message != '':
            raise OpensslError(error_message)

        # Return the decrypted content
        return result.stdout


class PublicKey:
    def __init__(self, public_key_path):
        self.public_key_path = public_key_path

    def encrypt(self, plaintext, cipher='aes-128-cbc'):
        # Use the public key to encrypt the plaintext
        args = ['openssl', 'rsautl', '-encrypt', '-' + cipher, '-pubin', '-inkey', self.public_key_path]
        result = subprocess.run(args, input=plaintext.encode('utf-8'), stdout=subprocess.PIPE, stderr=subprocess.PIPE)

        # If there is an error message on stderr, raise OpensslError
        error_message = result.stderr.decode()
        if error_message != '':
            raise OpensslError(error_message)

        # Return the encrypted content
        return result.stdout

ciphertext = "U2FsdGVkX1+Z7UpB7YmA/aDoccolES7oojW/ZvH6XHA7o3yOZO5Y+6ec3IUD2NgM\nkPKP4sVfv6PqCslx+m4opw=="
passphrase = "ISECR0XX"

try:
    plaintext = decrypt(ciphertext, passphrase)
    print("Decrypted content:", plaintext.decode('utf-8'))
except OpensslError as e:
    print("Error:", e)


```

```python
import subprocess
import binascii


def encrypt_pkey(msg, publickey):
    """
    Chiffre un message avec une clé publique
    paramètres: msg -> message à chiffrer
                publickey -> fichier contenant la clé publique
    """
    # Commande pour chiffrer un message avec la clé publique : openssl pkeyutl -encrypt -pubin -inkey <fichier contenant la pkey>
    args = ['openssl', 'pkeyutl', '-encrypt', '-pubin', '-inkey', publickey]

    # Si le message est une chaîne de caractères, le convertir en utf-8
    if isinstance(msg, str):
        msg = msg.encode('utf-8')

    # Exécute la commande pour chiffrer le message
    result = subprocess.run(
        args, input=msg, stdout=subprocess.PIPE, stderr=subprocess.PIPE)

    # Vérifie s'il y a des erreurs
    error_message = result.stderr.decode()
    if error_message != '':
        raise Exception(f"Erreur lors du chiffrement: {error_message}")

    # conversion en hexadécimal
    hex_encoded = binascii.hexlify(result.stdout).decode()
    return hex_encoded



if __name__ == "__main__":
                    # Message à chiffrer
    publickey = "publickey.pem"    # Fichier contenant la clé publique

    try:
        encrypted = encrypt_pkey("I got it!", publickey)
        print(f'message ecrypté en hexadécimal: \n{encrypted}')
    except Exception as e:
        print(e)
```
